```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
sns.set(style='darkgrid', palette='bright', font_scale=1.5)
```

```python
df = pd.read_csv('./StudentsPerformance.csv')
df.head()
```

```python
sns.pairplot(df[['math score', 'reading score', 'writing score']], height=4)
```

```python
# 수학 읽기 쓰기 점수 평균 구하기 
def average_score(dt):
    return(dt['math score'] + dt['reading score'] + dt['writing score']) / 3

df['average score'] = df.apply(average_score, axis=1)
```

```python
# 점심 식사 유무와 수학 점수 관계
sns.catplot(x='lunch', y='math score', hue='gender', kind='boxen', data = df,
           height=10, palette=sns.color_palette(['red', 'blue']))
plt.title('math')
```

```python
# 점심 식사 유무와 읽기 점수 관계
sns.catplot(x='lunch', y='reading score', hue='gender', kind='boxen', data = df,
           height=10, palette=sns.color_palette(['red', 'blue']))
plt.title('reading')
```

```python
# 점심 식사 유무와 쓰기 점수 관계
sns.catplot(x='lunch', y='writing score', hue='gender', kind='boxen', data = df,
           height=10, palette=sns.color_palette(['red', 'blue']))
plt.title('writing')
```

```python
# 선이수 과목 수행여부와 평균점수와의 관계
sns.catplot(x='test preparation course', y='average score', hue='gender', kind='boxen', data = df,
           height=10, palette=sns.color_palette(['red', 'blue']))
plt.title('average')
```

```python
# 부모의 학력과 평균점수와의 관계
sns.catplot(x='parental level of education', y='average score', kind='boxen', data = df,
           height=14)
plt.title('average')
plt.legend(loc='lower right')
# 왼쪽부터 학사 전문 석사 준학사 고등학교 고등학교
```
